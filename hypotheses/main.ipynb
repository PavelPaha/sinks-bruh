```python
from __future__ import annotations

from pathlib import Path
import shlex
import sys
import subprocess


def find_repo_root(start: Path | None = None) -> Path:
    p = (start or Path.cwd()).resolve()
    for _ in range(20):
        if (p / "hypotheses").exists() and (p / "scripts").exists():
            return p
        if p.parent == p:
            break
        p = p.parent
    raise RuntimeError(f"Repo root not found from cwd={Path.cwd()}")


repo_root = find_repo_root()

# ==== config (edit these) ====
label = "hallucinated"  # "hallucinated" | "incorrect" | "auto"
only = None  # e.g. "H5,H6" (or "H5_chat_sensitivity")
include_h7 = False
skip_missing = True

# ==== run all hypotheses ====
cmd = [sys.executable, str(repo_root / "hypotheses" / "analyze_all.py"), "--label", label]
if include_h7:
    cmd.append("--include_h7")
if only:
    cmd += ["--only", only]
if skip_missing:
    cmd.append("--skip_missing")

print("Running:", shlex.join(cmd))
subprocess.check_call(cmd, cwd=str(repo_root))

print("Done. See:")
print("-", repo_root / "hypotheses" / "H5_chat_sensitivity" / "plots")
print("-", repo_root / "hypotheses" / "H6_query_set_sensitivity" / "plots")

# Tip: run a single hypothesis directly (example):
# subprocess.check_call([sys.executable, str(repo_root / "hypotheses" / "H5_chat_sensitivity" / "scripts" / "run_all.py"), "--label", label], cwd=str(repo_root))
```
